```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.random.rand(100, 1)
true_w = 2.0
true_b = 1.0
noise = 0.1 * np.random.randn(100, 1) #some additional noise for the linear regression
y = true_w * X + true_b + noise
```

```python
w=np.random.randn(1,1)
b=np.random.randn(1)

learning_rate = 0.1
num_epochs = 1000
batch_size = 10 #taking it as 10 initially

def compute_gradients(X_batch, y_batch, w, b):

  n = X_batch.shape[0]
  y_pred = np.dot(X_batch, w) + b
  dw = (2/n)*np.dot(X_batch.T, y_pred - y_batch)
  db = (2/n)*np.sum(y_pred - y_batch)
  return dw, db
```

```python
loss_h = []

for epochs in range(num_epochs):
  indices = np.arange(X.shape[0])
  np.random.shuffle(indices)

  for i in range(0, X.shape[0], batch_size):
    batch_indices = indices[i:i+batch_size] #ensures randomness of the data
    X_batch = X[batch_indices]
    y_batch = y[batch_indices]

    dw, db = compute_gradients(X_batch, y_batch, w, b)

    w -= learning_rate * dw
    b -- learning_rate * db
  y_pred_f = np.dot(X, w) + b
  loss = np.mean((y_pred_f - y)**2) #mean squared error
  loss_h.append(loss)
  if epochs % 100 == 0:
    print(f'Epoch {epochs}, Loss: {loss:.4f}')

plt.figure(figsize=(8, 5))
plt.plot(range(num_epochs), loss_h, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.title('Loss Curve over Epochs')
plt.legend()
plt.grid(True)
plt.show()

```
